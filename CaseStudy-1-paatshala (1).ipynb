# Do the following operations on ‘cars_data’ dataset:

# 1. Read the dataset to the python environment.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('cars_data.csv')
```

```python
data
```

# 2. Check for the null values present in the dataset.

```python
data.isna().sum()  ##provides null value for each columns if it has any null values
```

```python
data.isna().sum().sum()   ## provides the total count of null values in a dataset
```

# 3. Plot a bar graph of male vs female buyers participated in the sales

```python
print("Values in each Gender \n",
data['Buyer Gender'].value_counts())
```

```python
sns.countplot(data=data,x='Buyer Gender')
plt.title("Bar graph of Male vs Female")
plt.show()
```

# Find the top 5 cars based on their sales price.


```python
data = data.sort_values(by=['Sale Price', 'Make'], ascending=False)
data.head(5) ##  to get top 5 cars based on sales price 
```

```python
data = data.sort_values(by=['Resell Price', 'Make'], ascending=True)
data.head(5) ##  to get least 5 cars based on Recell price 
```
